```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings("ignore")
sns.set()
```

```python
df = pd.read_csv('Mobiles Dataset (2025).csv',encoding="latin1")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#For cleaning numerical values
```

```python
def clean_numeric(column):
    return pd.to_numeric(column.str.replace(r'[^0-9.]', '', regex=True), errors='coerce')
```

```python
df
```

```python
price_columns = [
    'Launched Price (Pakistan)', 
    'Launched Price (India)', 
    'Launched Price (China)', 
    'Launched Price (USA)', 
    'Launched Price (Dubai)'
]

for col in price_columns:
    df[col] = clean_numeric(df[col])
```

```python
df
```

```python
df['Mobile Weight'] = clean_numeric(df['Mobile Weight'].str.replace('g', '', regex=False))
```

```python
df
```

```python
df['RAM'] = clean_numeric(df['RAM'].str.replace('GB', '', regex=False))
```

```python
df
```

```python
df['Battery Capacity'] = clean_numeric(df['Battery Capacity'].str.replace('mAh', '', regex=False))
df['Screen Size'] = clean_numeric(df['Screen Size'].str.replace(' inches', '', regex=False))
```

```python
df
```

```python
df.describe()
```

```python
#Distribution of mobile launches by year
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Launched Year', data=df, palette='viridis')
plt.title('Distribution of Mobile Launches by Year')
plt.xlabel('Launched Year')
plt.ylabel('Number of Models')

plt.show()
```

```python
#Average prices comparision across different countries
```

```python
avg_prices = df[price_columns].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_prices.plot(kind='barh', color='#2D336B')
plt.title('Average Mobile Prices Comparison by Country')
plt.xlabel('Average Price (in local currency)')

plt.show()
```

```python
#Top companies by number of models launched
```

```python
top_companies = df['Company Name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_companies.plot(kind='bar', color='#2D336B')
plt.title('Top Companies by Number of Models Launched')
plt.xlabel('Company Name')
plt.ylabel('Number of Models')

plt.show()
```

```python
#Battery Capacity vs Price Analysis
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Battery Capacity', y='Launched Price (USA)', data=df, color='#2D336B')
plt.title('Battery Capacity vs Price (USA)')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

```python
#Price vs RAM Analysis
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='RAM', y='Launched Price (India)', data=df, palette='Set2')
plt.title('Price vs. RAM (India)')
plt.xlabel('RAM')
plt.ylabel('Price (INR)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#Average Price by Top Companies
```

```python
top_company_prices = df[df['Company Name'].isin(top_companies.index)]
avg_price_by_company = top_company_prices.groupby('Company Name')['Launched Price (India)'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_company.index, y=avg_price_by_company.values, palette='Spectral')
plt.title('Average Price by Top Companies (India)')
plt.xlabel('Company Name')
plt.ylabel('Average Price (INR)')
plt.xticks(rotation=45)

plt.show()
```

```python
#Price Trends Over the Years
```

```python
plt.figure(figsize=(12, 6))
avg_price_by_year = df.groupby('Launched Year')['Launched Price (India)'].mean()
sns.lineplot(x=avg_price_by_year.index, y=avg_price_by_year.values, marker='o', color='#2D336B')
plt.title('Average Price Trends Over the Years (India)')
plt.xlabel('Launched Year')
plt.ylabel('Average Price (INR)')

plt.show()
```
